```python
query1='''
select
  year,
  month,
  count(Patent_Title) as total_patent
from (
  select
    CAST(LEFT(Grant_or_Publication_Date, 4) AS INTEGER) as year,
    SUBSTR(Grant_or_Publication_Date, 5, 2) as month,
    RIGHT(Grant_or_Publication_Date, 2) as date,
    Patent_Title
  from patents-public-data.uspto_oce_cancer.publications
)
where year between 2012 and 2023
and Patent_Title like '%data mining%'
group by year, month
order by year desc, month asc
'''

# Set up the query
df = client.query(query1).to_dataframe()


# dataframe
df
```
